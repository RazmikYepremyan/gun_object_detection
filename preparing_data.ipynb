```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
import os
```

```python
for i in os.listdir("labels"):
    with open(f"labels/{i}","r") as file:
        lines=file.readlines()[1:]
    with open(f"labels/{i}","w") as file:
        img=mpimg.imread(f"lmages/{i.split('.')[0]}.jpeg")
        h,w=img.shape[0],img.shape[1]
        print(h,w)
        for line in lines:
            xmin, ymin, xmax, ymax = map(float, line.strip().split())
            print(i,xmin,ymin,xmax,ymax)
            x_center = (xmin + xmax) / 2.0 / w
            y_center = (ymin + ymax) / 2.0 / h
            width = (xmax - xmin) / w
            height = (ymax - ymin) / h
            file.write(f"{0} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
```
